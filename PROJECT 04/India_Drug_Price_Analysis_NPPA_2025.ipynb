# 🏥 India Drug Price Analysis — NPPA DPCO 2025
### Source: National Pharmaceutical Pricing Authority (NPPA) — S.O. 1489(E), Effective 01-April-2025
**Dataset:** 299 price-controlled drug formulations | **Framework:** Drugs (Prices Control) Order (DPCO) 2013 | **WPI Adjustment:** 1.74028% for 2025

---


## Phase 1 — Dataset Collection
Data sourced from NPPA's official ceiling price notification under DPCO 2013 (NLEM 2022 basis). 
The NPPA fixes ceiling prices annually based on the Wholesale Price Index (WPI). All 299 formulations in this dataset are **legally binding price caps** — manufacturers cannot sell above these prices.


## Phase 2 — Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline

df = pd.read_csv("drug_prices_india.csv")
print(f"Shape: {df.shape}")
print(f"
Missing values:
{df.isnull().sum()}")
df = df.dropna()
df["Ceiling_Price_INR"] = pd.to_numeric(df["Ceiling_Price_INR"], errors="coerce")
df = df.dropna(subset=["Ceiling_Price_INR"])
df["Drug_Name"] = df["Drug_Name"].str.strip().str.title()
print(f"
After cleaning: {len(df)} records")
df.head(10)
```

## Phase 3 — Exploratory Data Analysis

```python
print("=== PRICE STATISTICS (INR per unit) ===")
print(df["Ceiling_Price_INR"].describe())
print(f"
Median: ₹{df["Ceiling_Price_INR"].median():.2f}")
print(f"Mean:   ₹{df["Ceiling_Price_INR"].mean():.2f}")
print(f"Max:    ₹{df["Ceiling_Price_INR"].max():,.2f} ({df.loc[df["Ceiling_Price_INR"].idxmax(),"Drug_Name"]})")
print(f"Min:    ₹{df["Ceiling_Price_INR"].min():.2f} ({df.loc[df["Ceiling_Price_INR"].idxmin(),"Drug_Name"]})")
```

```python
print("=== AVERAGE PRICE BY THERAPEUTIC CATEGORY ===")
cat_stats = df.groupby("Therapeutic_Category")["Ceiling_Price_INR"].agg(["mean","median","max","count"])
cat_stats.columns = ["Mean (₹)","Median (₹)","Max (₹)","Count"]
cat_stats = cat_stats.sort_values("Mean (₹)", ascending=False)
cat_stats["Mean (₹)"] = cat_stats["Mean (₹)"].map("₹{:,.2f}".format)
cat_stats["Median (₹)"] = cat_stats["Median (₹)"].map("₹{:,.2f}".format)
cat_stats["Max (₹)"] = cat_stats["Max (₹)"].map("₹{:,.2f}".format)
cat_stats
```

```python
print("=== TOP 15 MOST EXPENSIVE FORMULATIONS ===")
top15 = df.nlargest(15, "Ceiling_Price_INR")[["Drug_Name","Strength","Therapeutic_Category","Ceiling_Price_INR"]]
top15["Ceiling_Price_INR"] = top15["Ceiling_Price_INR"].map("₹{:,.2f}".format)
top15.reset_index(drop=True, index_col=False)
display(top15.reset_index(drop=True))
```

## Phase 5 — Business Insights

### 🔬 Key Findings as a Healthcare Consultant

#### 1. The Price Distribution is Highly Right-Skewed
Over 75% of NLEM-regulated drugs are priced below ₹25/unit, making essential medicines genuinely affordable. However, specialty drugs (anti-cancer biologics, immunoglobulins) reach tens of thousands of rupees per dose, indicating that affordability challenges are concentrated in specific disease areas.

#### 2. Anti-Cancer Drugs Dominate the High-Price Segment
With an average ceiling price of **₹2,661/unit** and outliers like Trastuzumab (₹54,750/vial), oncology represents the greatest affordability challenge. While NPPA has capped prices, a full course of cancer treatment can still cost lakhs of rupees. Government procurement under Jan Aushadhi and hospital formulary schemes can dramatically reduce patient burden.

#### 3. Injections Carry Disproportionate Cost
Injections average **₹1,263/unit** vs ₹11 for tablets. This reflects the biological complexity and cold-chain requirements of injectable formulations. However, it also signals that hospital-administered care is significantly more expensive, which has implications for healthcare infrastructure investment.

#### 4. Anti-Infective Drugs are Well Covered — and Affordable
With 56 formulations and a median of just ₹16.76/unit, antibiotics, antivirals, and antifungals are broadly accessible. This reflects India's strategic prioritization of infectious disease control. Brand-to-generic price gaps in this segment offer policy leverage.

#### 5. Cardiovascular Drugs Show Strong DPCO Control
Despite cardiovascular disease being India's leading cause of mortality, drugs like Amlodipine (₹1.82–5.54/tablet), Atorvastatin (₹5–41/tablet), and Aspirin (₹0.36/tablet) remain accessible. This is a policy success story — long-term prevention therapy is financially viable for most patients.

#### 6. Generic Substitution Opportunity is Largest in Anti-Cancer & Anti-Fungal
Liposomal Amphotericin B (₹6,764/vial) and Trastuzumab (₹54,750/vial) are areas where biosimilar development and generic entry can save patients and the healthcare system billions. India's pharmaceutical manufacturing strength positions domestic companies to compete in these segments.

#### 7. NPPA Regulation Covers 23 Therapeutic Areas
The breadth of the DPCO list signals comprehensive policy intent. However, non-scheduled drugs (those outside NLEM) still carry uncapped prices — a gap that patients with non-essential but important conditions may face.

### 📌 Strategic Recommendations
1. **Biosimilar Investment:** Prioritize local biosimilar development for Trastuzumab, Bortezomib, and Bendamustine
2. **Jan Aushadhi Scale-Up:** Expand generic dispensing for high-volume cardiovascular and anti-diabetic drugs where MRP-ceiling gaps are largest
3. **Insurance Product Design:** Separate insurance products for oncology & immunology, given the concentration of high-cost drugs in these categories
4. **Hospital Procurement Strategy:** Leverage DPCO prices for institutional bulk purchasing tenders to achieve sub-ceiling rates
